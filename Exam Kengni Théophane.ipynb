```python
import matplotlib.pyplot as plt 
import numpy
```

```python
def draw_my_function(f, start, end, number_of_points):
    x_values = numpy.linspace(start, end, number_of_points)
    y_values = f(x_values)
    
    df_dx = numpy.gradient(y_values, x_values)
    
    plt.figure(figsize=(15,6)) # on instancie un graphe vide et on choisit sa taille (longueur, largeur)
    plt.plot(x_values, y_values, label='f(x)')
    plt.plot(x_values, df_dx, label="df(x)/dx")
    
    # Ajout de la légende et des titres
    plt.legend()
    plt.title('Représentation fonction f(x) et df(x)/dx')
    plt.xlabel('x')
    plt.ylabel('y')
    
    plt.show()
```

```python
#Question 1 : [0,10] 10000
```

```python
start = 0
end = 10
number_of_points = 10000

f= lambda x_values : numpy.exp(-x_values/10) * numpy.sin(x_values)
```

```python
draw_my_function(f, start, end, number_of_points)
```

```python
#Moyenne & écart type
```

```python
start = 0
end = 10
number_of_points = 10000

f= lambda x_values : numpy.exp(-x_values/10) * numpy.sin(x_values)

x_values = numpy.linspace(start, end, number_of_points)
y_values = f(x_values)
df_dx = numpy.gradient(y_values, x_values)

# Sélection des valeurs dans l'intervalle [3, 7] - Calcul de la moyenne et de l'écart type
s_indices = numpy.logical_and(x_values >= 3, x_values <= 7)
s_y_values = y_values[s_indices]
mean = numpy.mean(s_y_values)
std = numpy.std(s_y_values)

# Tracé de la fonction et de sa dérivée
plt.figure(figsize=(15,6))
plt.plot(x_values, y_values, label='f(x)')
plt.plot(x_values, df_dx, label="df(x)/dx")

# Tracé de la moyenne et de l'écart type
plt.axhline(mean, color='r', label='Moyenne')
plt.axhline(mean+std, linestyle='--', color='g', label='Moyenne + Ecart Type')
plt.axhline(mean-std, linestyle='--', color='y', label='Moyenne - Ecart Type')

# Ajout de la légende et des titres
plt.legend()
plt.title('Tracé de f(x) et df(x)/dx ave la moyenne et l\'écart type sur l\'intervalle [3, 7]')
plt.xlabel('x')
plt.ylabel('y')

plt.show()

print("Moyenne sur l'intervalle [3, 7] : ", mean)
print("Ecart-type sur l'intervalle [3, 7] : ", std)
```

```python
#f'(x) = 0 pour x £ [0,10]
```

```python
start = 0
end = 10
number_of_points = 10000

f= lambda x_values : numpy.exp(-x_values/10) * numpy.sin(x_values)

x_values = numpy.linspace(start, end, number_of_points)
y_values = f(x_values)
df_dx = numpy.gradient(y_values, x_values)

# Recherche des indices où la dérivée est proche de zéro
id_zero = numpy.where((df_dx) < 1e-2)

# Affichage des points où la dérivée est nulle
zero_x_values = x_values[id_zero]

# Tracé de la fonction et de sa dérivée
plt.figure(figsize=(15,6))
plt.plot(x_values, y_values, label='f(x)')
plt.plot(x_values, df_dx, label="df(x)/dx")

# Ajout de la légende et des titres
plt.legend()
plt.title('f\'(x) = 0 pour x £ [0,10]')
plt.xlabel('x')
plt.ylabel('y')

plt.show()

print("f\'(x) = 0 pour x £ [0,10]t :", zero_x_values)
```
